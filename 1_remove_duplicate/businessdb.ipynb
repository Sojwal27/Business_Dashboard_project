```python
# import pandas as pd

# # File paths
input_file = "/content/drive/MyDrive/business_database.csv"
output_file = "cleaned_business_data.csv"
chunk_size = 50000  # Process 500K rows at a time

# Use a set to track unique Company URLs
unique_urls = set()

# Open a file to store cleaned data
with pd.read_csv(input_file, chunksize=chunk_size, dtype=str) as reader:
    for i, chunk in enumerate(reader):
        # Drop duplicates within the chunk based on 'Company URL'
        chunk = chunk.drop_duplicates(subset=['Company URL'])

        # Filter out rows that have already been seen
        chunk = chunk[~chunk['Company URL'].isin(unique_urls)]

        # Add new URLs to the tracking set
        unique_urls.update(chunk['Company URL'].dropna())

        # Append to the cleaned CSV file
        chunk.to_csv(output_file, mode='a', index=False, header=(i == 0))  # Write header only once

print(f"Cleaned data saved to {output_file}")
```
