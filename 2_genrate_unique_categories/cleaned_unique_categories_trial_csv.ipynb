```python
import pandas as pd

chunk_size = 100000  # Adjust based on your available memory
chunks = []

for chunk in pd.read_csv("business_collection.csv", dtype=str, chunksize=chunk_size, low_memory=False):
    chunks.append(chunk)

df = pd.concat(chunks)
```

```python
df.head()
```

```python
df = df.drop(columns=['Contact Person'])
```

```python
print(df.columns)
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
# Verify the column is gone
print(df.columns)  # Should NOT include 'Contact Person'
```

```python
# Convert all columns to string
df = df.astype(str)
```
